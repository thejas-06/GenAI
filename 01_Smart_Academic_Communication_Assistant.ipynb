# Import library and create client

```python
from google import genai
from getpass import getpass

# Enter your Gemini API key safely
api_key = getpass("Enter your Gemini API Key: ")

# Create client
client = genai.Client(api_key=api_key)
```

# Get user input

```python
print("SMART ACADEMIC COMMUNICATION ASSISTANT")
print("-" * 50)

communication_type = input("Enter communication type (Notice / Email / WhatsApp Message): ")
purpose = input("Enter the purpose: ")
audience = input("Enter the target audience: ")
key_points = input("Enter the important points (comma separated): ")
tone = input("Enter the tone (Formal / Professional / Friendly / Concise): ")
length = input("Enter the length (Short / Medium / Detailed): ")
```

# Build the prompt

```python
prompt = f'''
You are an academic communication assistant.

Generate a {communication_type} for the following requirement.

Purpose:
{purpose}

Target Audience:
{audience}

Important Points:
{key_points}

Tone:
{tone}

Length:
{length}

Instructions:
1. The content must be clear and grammatically correct.
2. Keep it suitable for academic or official use.
3. Use proper formatting.
4. Do not add unnecessary information.
5. Make the message natural and professional.
'''
```

# Generate the output

```python
response = client.models.generate_content(
model="gemini-2.5-flash",
contents=prompt
)

print("\nGENERATED OUTPUT")
print("-" * 50)
print(response.text)
```
